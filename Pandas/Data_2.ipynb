# Extract 10 Year ridership average for each station

```python
max_date = cta_monthly_df_reorg['date'].max()
max_date
```

```python
ten_yrs_ago = max_date - relativedelta(years=11)
ten_yrs_ago
```

```python
cta_monthly_ten_years_data = cta_monthly_df_reorg[cta_monthly_df_reorg['date'] > ten_yrs_ago]

cta_monthly_ten_years_data.head()
```

```python
cta_monthly_ten_years_data['date'] = pd.to_datetime(cta_monthly_ten_years_data['date']).dt.to_period('y')
cta_monthly_ten_years_data.head()
```

```python
ten_year_ridership_data = cta_monthly_ten_years_data.pivot_table(index=['Station_ID','Station_Name'], 
                    columns='date', 
                    values='monthtotal', 
                    aggfunc='sum')
ten_year_ridership_data.head()
```

```python
ten_year_ridership_data = ten_year_ridership_data.reset_index().rename_axis(None).rename_axis(None, axis=1)

ten_year_ridership_data.head()
```

```python
ten_year_data_sorted = ten_year_ridership_data.sort_values('Station_Name')
ten_year_data_sorted.head(10)
```

```python
ten_year_data_sorted.count()
```

```python
ten_year_data_merged = pd.merge(ten_year_data_sorted, system_stations_df, on="Station_ID")
ten_year_data_merged.head()
```

```python
ten_year_data_unique = ten_year_data_merged.drop_duplicates(subset='Station_ID', keep="first", inplace=False)
ten_year_data_unique.tail(10)
```

```python
ten_year_data_unique.ADA.replace(['True', 'False'], ['Accessible','Not Accessible'], inplace=True)
ten_year_data_unique['Red'] = ten_year_data_unique['Red'].astype('bool')
ten_year_data_unique['Blue'] = ten_year_data_unique['Blue'].astype('bool')
ten_year_data_unique['Green'] = ten_year_data_unique['Green'].astype('bool')
ten_year_data_unique['Brown'] = ten_year_data_unique['Brown'].astype('bool')
ten_year_data_unique['Purple'] = ten_year_data_unique['Purple'].astype('bool')
ten_year_data_unique['Purple_Express'] = ten_year_data_unique['Purple_Express'].astype('bool')
ten_year_data_unique['Yellow'] = ten_year_data_unique['Yellow'].astype('bool')
ten_year_data_unique['Orange'] = ten_year_data_unique['Orange'].astype('bool')
ten_year_data_unique['Pink'] = ten_year_data_unique['Pink'].astype('bool')
```

```python
ten_year_data_unique.head()
```

```python
ten_year_data_unique.dtypes
```

```python
ten_year_data = ten_year_data_unique#.drop(columns=['Station_Name_old'])
ten_year_data.tail(12)
```

```python
ten_year_data['Station_ID'].nunique()
```
